```python
# Importing required libraries
import pandas as pd          # For data loading & data manipulating
import matplotlib.pyplot as plt          # For data visualization(plots and graphs)              
from sklearn.linear_model import LogisticRegression          # For building a Logistic Regression machine learning model
```

```python
# Loading the Iris dataset from the CSV file into a DataFrame
iris_data=pd.read_csv('iris.csv') 
iris_data.head()          # Displaying the first 5 rows of the dataset to quickly understand its structure
```

```python
# Displaying the last 5 rows of the Iris dataset
iris_data.tail()
```

```python
# Show summary information about the Iris dataset (columns, data types, null values)
iris_data.info()
```

```python
# Plot sepal length vs width for each species
plt.figure(figsize=(6,4))
for species in iris_data['Species'].unique():
    subset = iris_data[iris_data['Species'] == species]
    plt.scatter(subset['SepalLengthCm'], subset['SepalWidthCm'], label=species)

# Labels and display with legend
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Width")
plt.legend()
plt.show()
```

```python
# Plot petal length vs width for each species
plt.figure(figsize=(6,4))
for species in iris_data['Species'].unique():
    subset = iris_data[iris_data['Species'] == species]
    plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'], label=species)

# Labels and display with legend
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Width")
plt.legend()
plt.show()
```

```python
# List of numeric columns to visualize
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']


for col in features:
    plt.figure(figsize=(6,4))          # Create a new plot for each feature
    plt.hist(iris_data[col], bins=12)  # Plot distribution of the feature
    plt.title(f"Histogram of {col}")   # Set title for clarity
    plt.xlabel(col)
    plt.ylabel("Frequency")            # Y-axis label (count of values)
    plt.show()
```

```python
# Create a boxplot to compare feature distributions
plt.figure(figsize=(8,5))
iris_data[features].plot(kind='box')          # Boxplot for all selected numeric columns
plt.title("Boxplot of Iris Features")         # Title for better understanding
plt.show()
```

```python
# Split the dataset into features (X) and target labels (y)
X=iris_data.drop(columns=['Id','Species'])          # X = all columns except Id and Species
y=iris_data['Species']                              # Y = species column (target)
```

```python
# Create a Logistic Regression model
model=LogisticRegression()
```

```python
# Train the logistic regression model using X (features) and y (labels)
model.fit(X,y)
```

```python
# Create a sample input for prediction with the same feature columns
sample=pd.DataFrame([[9.5,5,6,7]],columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
# Predict the species for the given sample input
prediction=model.predict(sample)
```

```python
# Display the predicted species for given Sample
print(prediction)
```
